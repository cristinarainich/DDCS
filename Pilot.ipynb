```python
pilot = pd.read_csv(
    working_path.joinpath('Pilot.csv'),
    index_col=None
)
```

```python
for j in range(1, 11):
  speed = []
  col1 = 'P' + str(j) + ' SYL'
  col2 = 'P' + str(j) + ' RT'
  for i in range(0, 8):
    x = pilot.iloc[i][col1] / (pilot.iloc[i][col2] * 60)
    speed.append(x)
  name = 'P' + str(j) + ' Speed'
  pilot[name] = speed
```

```python
pilot
```

```python
# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(
    pilot['P1 Error'].values, pilot['P2 Error'].values, pilot['P3 Error'].values,
    pilot['P4 Error'].values, pilot['P5 Error'].values, pilot['P6 Error'].values,
    pilot['P7 Error'].values, pilot['P8 Error'].values, pilot['P9 Error'].values,
    pilot['P10 Error'].values)

# Set your significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print("There is a significant difference between passages.")
else:
    print("There is no significant difference between passages.")

# Print the F-statistic and p-value
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
f_statistic, p_value = stats.f_oneway(
    pilot['P1 Speed'].values, pilot['P2 Speed'].values, pilot['P3 Speed'].values,
    pilot['P4 Speed'].values, pilot['P5 Speed'].values, pilot['P6 Speed'].values,
    pilot['P7 Speed'].values, pilot['P8 Speed'].values, pilot['P9 Speed'].values,
    pilot['P10 Speed'].values)

# Set your significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print("There is a significant difference between passages.")
else:
    print("There is no significant difference between passages.")

# Print the F-statistic and p-value
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```
